```python
% cd C:\Users\wolfg\Desktop\Porticode
import pandas as pd
import numpy as np
% matplotlib inline
```

```python
df = pd.read_csv("Dataset.csv")
```

```python
len(df)/250
```

```python
TIME_STEPS = 250
TIME_WINDOW = 10

sub_sub_list = []

for i in range(0,80,1): 
                #80
    
    sub_df = df[i*TIME_STEPS:(i+1)*TIME_STEPS] 
    
    stdevs = []
    means = []
    
    
    for k in range(1, 7, 1):
        
        result = sub_df.iloc[:, k].std()
        result_mean = sub_df.iloc[:, k].mean()
        
        stdevs.append(result)
        means.append(result_mean)
    
    sub_df['mean_x_acc_glob'] = means[0]
    sub_df['mean_y_acc_glob'] = means[1]
    sub_df['mean_z_acc_glob'] = means[2]
    sub_df['mean_x_gyr_glob'] = means[3]
    sub_df['mean_y_gyr_glob'] = means[4]
    sub_df['mean_z_gyr_glob'] = means[5]
     
    sub_df['stddev_x_acc_glob'] = stdevs[0]
    sub_df['stddev_y_acc_glob'] = stdevs[1]
    sub_df['stddev_z_acc_glob'] = stdevs[2]
    sub_df['stddev_x_gyr_glob'] = stdevs[3]
    sub_df['stddev_y_gyr_glob'] = stdevs[4]
    sub_df['stddev_z_gyr_glob'] = stdevs[5]
    
    result_list_x_acc = []
    result_list_y_acc = []
    result_list_z_acc = []
    result_list_x_gyr = []
    result_list_y_gyr = []
    result_list_z_gyr = []
    
    result_list_x_acc_mean = []
    result_list_y_acc_mean = []
    result_list_z_acc_mean = []
    result_list_x_gyr_mean = []
    result_list_y_gyr_mean = []
    result_list_z_gyr_mean = []    
    
    for x in range(0, len(sub_df)-TIME_WINDOW,1):
        
        sub_sub_df = sub_df[x:x+TIME_WINDOW]
        
        mean_x_acc = sub_sub_df['X (m/s^2)'].mean()
        mean_y_acc = sub_sub_df['Y (m/s^2)'].mean()
        mean_z_acc = sub_sub_df['Z (m/s^2)'].mean()
        mean_x_gyr = sub_sub_df['X (rad/s)'].mean()
        mean_y_gyr = sub_sub_df['Y (rad/s)'].mean()
        mean_z_gyr = sub_sub_df['Z (rad/s)'].mean()  
        
        std_x_acc = sub_sub_df['X (m/s^2)'].std()
        std_y_acc = sub_sub_df['Y (m/s^2)'].std()
        std_z_acc = sub_sub_df['Z (m/s^2)'].std()
        std_x_gyr = sub_sub_df['X (rad/s)'].std()
        std_y_gyr = sub_sub_df['Y (rad/s)'].std()
        std_z_gyr = sub_sub_df['Z (rad/s)'].std()
        
        result_list_x_acc_mean.append(mean_x_acc)
        result_list_y_acc_mean.append(mean_y_acc)
        result_list_z_acc_mean.append(mean_z_acc)
        result_list_x_gyr_mean.append(mean_x_gyr)
        result_list_y_gyr_mean.append(mean_y_gyr)
        result_list_z_gyr_mean.append(mean_z_gyr)
          
        result_list_x_acc.append(std_x_acc)
        result_list_y_acc.append(std_y_acc)
        result_list_z_acc.append(std_z_acc)
        result_list_x_gyr.append(std_x_gyr)
        result_list_y_gyr.append(std_y_gyr)
        result_list_z_gyr.append(std_z_gyr)
        
    sub_df = sub_df[0:240]
       
    sub_df['mean_x_acc_loc_w10'] = result_list_x_acc_mean
    sub_df['mean_y_acc_loc_w10'] = result_list_y_acc_mean
    sub_df['mean_z_acc_loc_w10'] = result_list_z_acc_mean
    sub_df['mean_x_gyr_loc_w10'] = result_list_x_gyr_mean
    sub_df['mean_y_gyr_loc_w10'] = result_list_y_gyr_mean
    sub_df['mean_z_gyr_loc_w10'] = result_list_z_gyr_mean    
        
    sub_df['std_x_acc_loc_w10'] = result_list_x_acc
    sub_df['std_y_acc_loc_w10'] = result_list_y_acc
    sub_df['std_z_acc_loc_w10'] = result_list_z_acc
    sub_df['std_x_gyr_loc_w10'] = result_list_x_gyr
    sub_df['std_y_gyr_loc_w10'] = result_list_y_gyr
    sub_df['std_z_gyr_loc_w10'] = result_list_z_gyr
    
    sub_sub_list.append(sub_df)
```

```python
monster_df = pd.concat(sub_sub_list)
```
